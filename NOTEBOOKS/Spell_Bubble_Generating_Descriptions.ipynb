```python
import json
import openai

# Load the images in base64
with open('imagens_base64.json', 'r') as file:
    imagens_base64 = json.load(file)

# Load JSON information
with open('associacoes.json') as file:
    associacoes = json.load(file)

# Select only the first panel
panel_key = "painel1"
base64_image = imagens_base64[panel_key]
assoc_info = associacoes[panel_key]

prompt_text = (
    "Create a clear and concise description of the comic panel for a visually impaired audience.\n"
    "Begin by identifying the panel type (e.g., horizontal, vertical, square) and include any captions.\n"
    "Then, describe the scene's setting based only on visible elements in the foreground. Try to understand the daytime based on contrast.\n"
    "Avoid making assumptions about backgrounds. If something is indistinguishable or unusual, don't mention it.\n"
    "After, describe the characters (human or not) as they appear in the panel. Know that Whiz is a bird, not a human.\n"
    "Describe characters' actions and appearance accurately. Don't over-describe the costumes, don't judge or invent.\n"
    "Refrain from interpreting facial expressions, gestures, and body postures.\n"
    "Include characters' dialogues in the middle of the narrative exactly as they appear in the 'associacoes.json' file.\n"
    "For characters present in the panel without dialogue, simply mention their presence. The names are provided in the 'associacoes.json' file.\n"
    "Lastly, describe significant graphic elements such as shadows, bright or deep colors, and patterns.\n"
    "Avoid adding historical context or speculative details about the characters or scene.\n"
    "Strictly avoid inferring or imagining details not present in the image or the 'associacoes.json' file.\n\n"
)

prompt = {
    "role": "user",
    "content": [
        {"type": "text", "text": prompt_text},
        {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{base64_image}", "detail": "high"}},
        {"type": "text", "text": f"See the file 'associacoes.json': {json.dumps(assoc_info)} to know the characters' names, their positioning in the panel and to have context about the scene based on the speeches. When using this data for context, trust the given information. Use the names and dialogues, but don't mention the file source in the transcription."}
    ]
}

# Send the prompt to the API and process the response
response = client.chat.completions.create(
    model="gpt-4-vision-preview",
    messages=[prompt],
    max_tokens=350
)

# Print the description of the first panel
print(f"Painel 1 Descrição:\n{response.choices[0].message.content}\n")
```

```python
# Storing the panel description
painel_descricao = response.choices[0].message.content

# Create a dictionary to save the result
resultado = {
    panel_key: painel_descricao
}

# Save the result to a JSON file
with open('descricao_painel1.json', 'w') as json_file:
    json.dump(resultado, json_file)

print(f"Painel 1 Descrição salva com sucesso:\n{painel_descricao}\n")
```

# Testing prompts in sequential storytelling

```python
import json
import openai

# Load images in base64 and JSON information
with open('imagens_base64.json', 'r') as file:
    imagens_base64 = json.load(file)

with open('associacoes.json') as file:
    associacoes = json.load(file)

# Dictionary to store panel descriptions
descricao_paineis = {}

# Iterate over the first and second panels
for idx, panel_key in enumerate(associacoes.keys()):
    # Limit the iteration only to the first two panels
    if idx > 1:
        break

    base64_image = imagens_base64[panel_key]
    assoc_info = associacoes[panel_key]

    # Check if it's the first panel or subsequent panels
    if idx == 0:
        setting_instruction = "2. Describe the scene's setting, including the view type (e.g., wide view, mid view, close up, etc.). Consider the daytime based on contrast (if possible), and avoid guessing or interpreting unusual background/foreground elements.\n"
        character_description_instruction = "3. Describe the characters' actions, appearance, and visibility (e.g., full figure, from the knees up, neck up, etc.), accurately. Know that Whiz is a bird, not a human. Don't over-describe the costumes. Be cautious over colors. Never be judgemental or inventive.\n"
        graphic_elements_instruction = "8. Describe significant graphic elements such as shadows, bright or deep colors, and patterns, only if they are particularly noticeable.\n"
    else:
        setting_instruction = "2. Briefly acknowledge the scene's setting and focus on new or changed elements since the previous panel. Include the view type if it contributes to the narrative or has changed from the previous panel.\n"
        character_description_instruction = "3. Continuing the narrative from the previous panels, focus on new actions, changes in characters' appearance, visibility or new characters introduced. Avoid reintroducing known characters in detail. Know that Whiz is a bird, but don't mention his physical form.\n"
        graphic_elements_instruction = "8. Describe significant graphic elements, such as shadows, bright or deep colors, and patterns, only if there is a considerable change from the previous panel.\n"

    # Define the prompt text
    prompt_text = (
    "Create a clear and concise structured description of comic panel {}. The description should be accessible and engaging for a visually impaired audience. Keep the description narrative-focused, continuing the story from the previous panels.\n\n"
    "1. Begin by identifying the panel type (e.g., vertical, horizontal, square, etc.) and include any captions. If there is no narrator's caption, don't mention it.\n" +
    setting_instruction +
    character_description_instruction +
    "4. In the characters' description, always mention the names of the characters present in the panel, which are provided in the 'associacoes.json' file. Mention their presence even if they don't have a speech line. Do not interpret or explain the characters' presence.\n"
    "5. Include characters' dialogue exactly as they appear in the 'associacoes.json' file in the middle of the narrative. Never anticipate, explain, or interpret the content and the tone of the speeches; know the readers are not dumb and shall not receive spoilers.\n"
    "6. Refrain from interpreting facial expressions, gestures, and body postures. Focus on elements that contribute meaningfully to the narrative.\n"
    "7. In scenes with multiple characters, especially in active or tense scenes, prioritize the flow of the narrative over detailed figure description.\n" +
    graphic_elements_instruction +
    "9. Never add historical context or speculative details about the characters or scene. Never interpret, explain, or anticipate the moment of any scene.\n"
    "10. Strictly avoid inferring or imagining details not present in the image or the 'associacoes.json' file.\n\n"
    "The structure must be divided into clear sections, mixing the dialogue with the characters' descriptions. Never interpret, explain, conclude, summarize, or anticipate the narrative flow. Never add extra sections, comments, speculations, or observations besides what was asked in this prompt. Don't describe what commands you don't do/did.\n\n"
    ).format(idx + 1)

    # Create and send the prompt
    prompt = {
        "role": "user",
        "content": [
            {"type": "text", "text": prompt_text},
            {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{base64_image}", "detail": "high"}},
            {"type": "text", "text": f"See the file 'associacoes.json': {json.dumps(assoc_info)} to know the characters' names, their positioning in the panel and to have context about the scene based on the speeches. When using this data for context, trust the given information. Use the names and dialogue in the description, but never mention the source."}
        ]
    }

    response = client.chat.completions.create(
        model="gpt-4-vision-preview",
        messages=[prompt],
        max_tokens=500
    )

    # Store the panel description
    descricao_paineis[panel_key] = response.choices[0].message.content

    # Print the panel description
    print(f"Painel {idx + 1} Descrição:\n{response.choices[0].message.content}\n")
```

```python
# After the loop, save the 'panel_descriptions' dictionary to a JSON file
with open('descricao_paineis.json', 'w') as json_file:
    json.dump(descricao_paineis, json_file)

print(f"Descrições salvas com sucesso:\n{descricao_paineis}\n")
```
